# YOLOv8的学习

```python
from ultralytics import YOLO
```

### 使用现成的模型

```python
model = YOLO('yolov8n.pt') # 直接下载权重，使用框架

result = model(source = 0, show = True, conf = 0.4, save = False)
# source 能使用图片，视频等，0表示摄像头，save表示是否保存
```

```python
import cv2
import os
model_1 = YOLO('yolov8n-seg.pt')
video_path = "事故.mp4"
file_path = 'img_temporarily/'
cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
frame_idx = 0

while cap.isOpened():
    success, frame = cap.read()
    if success:
        result = model_1(frame)
        annotated_result = result[0].plot()
        cv2.imshow('frame', annotated_result)
        cv2.imwrite(f'img_temporarily/{str(frame_idx)}.jpg', annotated_result)
        frame_idx += 1
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break


out_path = '事故框.mp4'
img_list = os.listdir(file_path)
img_list.sort()
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(out_path, fourcc, fps, (frame_width, frame_height))

for img_name in img_list:
    img_path = os.path.join(file_path, img_name)
    frame = cv2.imread(img_path)
    out.write(frame)

out.release()
cap.release()
cv2.destroyAllWindows()

def delete(path):
    for _file_name in os.listdir(path):
        os.remove(os.path.join(path, _file_name))

delete(file_path)
```
